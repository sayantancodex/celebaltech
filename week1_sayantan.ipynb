### 1.3 Exceptions

Write `safe_divide(a, b)` that:
- Returns `a / b` normally
- Returns `None` if `b == 0` (catch `ZeroDivisionError`)
- Raises `TypeError` with message `"Inputs must be numeric"` if either input is not a number

````python
def safe_divide(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Inputs must be numeric")
    try:
        return a / b
    except ZeroDivisionError:
        return None

# Tests
assert safe_divide(10, 2) == 5.0
assert safe_divide(5, 0) is None
try:
    safe_divide('x', 2)
    assert False, "Should have raised TypeError"
except TypeError as e:
    assert str(e) == "Inputs must be numeric"
print("1.3 passed ✓")
````

### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lambda_val = eigenvalues[i]
    Av = A @ v
    lambda_v = lambda_val * v
    assert np.allclose(Av, lambda_v), f"Verification failed for eigenpair {i}"

# Plot
fig, ax = plt.subplots(figsize=(6, 6))
origin = [0, 0]

# Plot original matrix A, for context
ax.quiver(*origin, A[0, 0], A[1, 0], color='lightgray', scale=1, units='xy', width=0.02, label='Column 1 of A')
ax.quiver(*origin, A[0, 1], A[1, 1], color='silver', scale=1, units='xy', width=0.02, label='Column 2 of A')

# Plot eigenvectors scaled by eigenvalues
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lambda_val = eigenvalues[i]
    ax.quiver(*origin, v[0] * lambda_val, v[1] * lambda_val, angles='xy', scale_units='xy', scale=1, color=f'C{i}', label=f'Eigenvector {i+1} (scaled)')
    ax.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=f'C{i}', linestyle=(0, (1, 1)), alpha=0.5, label=f'Eigenvector {i+1} (unscaled)') # Changed linestyle to a minimal tuple pattern

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Eigenvectors scaled by Eigenvalues')
ax.legend()
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

> For a given matrix, an **eigenvector** is a special vector that, when transformed by the matrix, only changes in magnitude (is stretched or shrunk) but not in direction. The **eigenvalue** is the scalar factor by which the eigenvector is stretched or shrunk. Geometrically, if a matrix 'stretches' a vector, it means that the transformation applied by the matrix scales the vector along its original direction without rotating it. Eigenvectors represent the directions along which the transformation is purely a scaling operation. This is crucial for understanding linear transformations, principal component analysis, and many other areas in machine learning.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # Create bins
    min_val = min(expected.min(), actual.min())
    max_val = max(expected.max(), actual.max())
    bins = np.linspace(min_val, max_val, bins + 1)

    # Calculate counts in each bin
    expected_counts = np.histogram(expected, bins=bins)[0]
    actual_counts = np.histogram(actual, bins=bins)[0]

    # Calculate percentages
    expected_pct = expected_counts / len(expected) + 1e-10 # Add epsilon
    actual_pct = actual_counts / len(actual) + 1e-10       # Add epsilon

    # Calculate PSI
    psi = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
sns.histplot(train_dist, kde=True, color='blue', label='Expected (Train)', ax=ax, stat='density', alpha=0.5)
sns.histplot(drift_dist, kde=True, color='red', label='Actual (Drift)', ax=ax, stat='density', alpha=0.5)
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Interpretation (5.4):**

- **Kolmogorov-Smirnov Test:**
  - For `s1` (generated from Normal distribution): The p-value (0.6268) is > 0.05, so we fail to reject the null hypothesis. This means `s1` is likely drawn from a normal distribution, as expected.
  - For `s2` (generated from Exponential distribution): The p-value (0.0000) is < 0.05, so we reject the null hypothesis. This means `s2` is *not* likely drawn from a normal distribution, also as expected.

- **Augmented Dickey-Fuller Test:**
  - **Original Series `ts`:** The p-value (0.7802) is > 0.05, indicating that the original time series is non-stationary (it has a unit root, consistent with a linear trend).
  - **Differenced Series `ts_diff`:** After differencing, the p-value (0.0000) is < 0.05, leading us to reject the null hypothesis. This confirms that the differenced series is now stationary, as the linear trend has been removed.

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** Occurs when the relationship between the input variables and the target variable changes over time. The 'concept' or underlying phenomenon the model is trying to predict changes.
- **Covariate drift:** Occurs when the distribution of the input features (covariates) changes over time, but the relationship between the inputs and the target remains the same.
- **PSI < 0.1:** Indicates no significant shift or a minor change in distribution. The model's performance is likely stable due to this feature.
- **PSI 0.1–0.2:** Indicates a moderate shift in distribution. This might warrant further investigation or closer monitoring.
- **PSI > 0.2:** Indicates a significant or major shift in distribution. This typically suggests that the model's feature distribution has changed substantially, and retraining the model or investigating the data quality is recommended.
- **Retraining trigger example:** If the PSI for a critical input feature exceeds 0.2 for three consecutive days, it could automatically trigger a model retraining process to adapt to the new data distribution.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red   = red / total
p_blue  = blue / total
p_green = green / total

# Joint: P(red first AND blue second)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional: P(blue second | first red)
p_blue_given_red = blue / (total - 1)

# Independence check: are P(B|A) == P(B) ?
# P(B) = P(second is blue). This can be calculated as the sum of:
# P(first red, second blue) + P(first blue, second blue) + P(first green, second blue)
p_second_blue = (red / total) * (blue / (total - 1)) + \
                (blue / total) * ((blue - 1) / (total - 1)) + \
                (green / total) * (blue / (total - 1))

# Independent should be True if P(B|A) == P(B), False otherwise
independent = np.isclose(p_blue_given_red, p_second_blue)

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_second_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)
    posterior_spam = (p_word_given_spam * prior_spam) / p_word
    return posterior_spam

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The initial probability that any given email is spam (0.30) |
| Likelihood | P(word \| Spam) | The probability of the word "free" appearing in a spam email (0.80) |
| Evidence | P(word) | The overall probability of the word "free" appearing in any email (0.2750) |
| Posterior | P(Spam \| word) | The updated probability that an email is spam, given that it contains the word "free" (0.8727) |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, sample_size, replace=True).mean() for _ in range(n_samples)])

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
sns.histplot(sample_means, kde=True, stat='density', ax=ax, color='skyblue', label='Sample Means Distribution')

# Overlay theoretical normal distribution
x_clt = np.linspace(sample_means.min(), sample_means.max(), 100)
pdf_clt = stats.norm.pdf(x_clt, loc=pop_mean, scale=clt_std)
ax.plot(x_clt, pdf_clt, color='red', linestyle='--', label='Theoretical Normal (CLT)')

ax.legend()
ax.set_xlabel('Sample Mean')
ax.set_ylabel('Density')
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> The Central Limit Theorem is fundamental because it states that the distribution of sample means will be approximately normal, regardless of the population's distribution, as long as the sample size is sufficiently large. This result is crucial for both statistics and machine learning as it justifies the use of parametric statistical tests (like t-tests) and confidence intervals on sample means, even when the underlying data is not normally distributed. In machine learning, it underpins assumptions in various algorithms and the understanding of error distributions.
